#### 1. Write a Python Program to implement your own myreduce() function which works exactly like Python's built-in function reduce()

```python
def my_reduce(a,b):
    total = 0
    for i in range(a, b):
        total = total + i
    return total
```

```python
print("My reduce function output: ",my_reduce(0,11))
```

```python
#checking against the python reduce function

def add(x, y):
    return x+y

from functools import reduce
output  = reduce(add, range(0,11))
print("Python reduce function output: ", output)
```

##### 2. Write a Python program to implement your own myfilter() function which works exactly like Python's built-in function filter()

```python
#defining a function to test filter
def even_check(num):
    if num%2 ==0:
        return True
```

```python
def my_filter(a,b):
    lst = []
    for i in range(a,b):
        if i % 2 == 0:
            lst.append(i)
    return lst
```

```python
my_filter(1,9)
```

```python
# Checking aginst the python filter function

lst =[1,2,3,4,5,6,7,8]
list(filter(even_check,lst))
```
